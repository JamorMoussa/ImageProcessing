```python
def AddImages(img,im2,x,y,alpha):
  w,h = im2.shape[:-1]
  x -= w//2
  y -= h//2
  if w < img.shape[0] and h < img.shape[1]:
      for i in range(w):
        for j in range(h//3):
          if i+x-2 > 0 and i+x < img.shape[0] and (y + (j+1)*3 -2) > 0 and (y + (j+1)*3) < img.shape[1]:
            img[i+x,y + j*3: y + (j+1)*3] = (1-alpha)*img[i+x,y + j*3: y + (j+1)*3] + alpha*im2[i,j*3: (j+1)*3]
  else:
    raise Exception('Error')
  return img
```

```python
dst = im1.copy()
dst = AddImages(dst,im2,100,100,0.2)
plt.imshow(dst)
```
